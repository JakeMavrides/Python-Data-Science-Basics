```python
# https://pandasguide.readthedocs.io/en/latest
import pandas as pd
# https://matplotlib.org/stable/tutorials/introductory/pyplot.html
import matplotlib.pyplot as plt
```

```python
# a simple line plot (0,1) (1,2) ... (9,10)
yvals = list(range(1,10+1))
plt.plot(yvals)
plt.show()
```

```python
# not so linear...
yvals = [0.10, 0.20, 0.40, 0.50, 0.65, 0.75, 0.70, 0.60, 0.40, 0.35, 0.20, 0.10]
plt.plot(yvals)
plt.show()
```

```python
# adding some labels
plt.plot(yvals)
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
# adding some color (red -> 'r') and the shape 'x' for f(x) = x^2
# blue in 'o' for g(x) = 2.5*x
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25], 'rx', label='f(x)')
plt.plot([1, 2, 3, 4, 5], [2.5, 5, 7.5, 10, 12.5], 'bo', label='g(x)')
plt.legend()
plt.show()
```

```python
# labels
names = ['label_a', 'label_b', 'label_c']
values = [10, 40, 100]

# set the dimensions of the overall figure
plt.figure(figsize=(12, 3))

# 1 x 3 subplot, entry 1 -> bar graph
plt.subplot(131)
plt.bar(names, values)
# 1 x 3 subplot, entry 2 -> scatterplot
plt.subplot(132)
plt.scatter(names, values)
# 1 x 3 subplot, entry 3 -> line plot
plt.subplot(133)
# x-axis = names, y-axis = values | also, thick line
plt.plot(names, values, linewidth=3.0)
# set the overall title
plt.suptitle('Categorical Plotting')
# show all plots
plt.show()
```

```python
# source: https://www.kaggle.com/datasets/zaraavagyan/weathercsv
wdf = pd.read_csv('sample_data/weather.csv')
print(wdf.shape)
wdf.sample(10)
```

```python
plt.figure(figsize=(12, 3))
plt.plot(wdf['MinTemp'])
plt.xticks([])
plt.xlabel('Days over Time')
plt.ylabel('MinTemp (in Celcius)')
plt.show()
```

```python
# create dataframe 'wdf_nrt' to represent the days where it did NOT rain
wdf_nrt = wdf[wdf['RainToday'] == 'No']
# create dataframe 'wdf_rt' to represent the days where it did rain
wdf_rt = wdf[wdf['RainToday'] == 'Yes']
print(wdf_nrt.shape)
print(wdf_rt.shape)
wdf_nrt.sample(5)
```

```python
names = ['No Rain Today','Rain Today']
plt.figure(figsize=(6, 3))
# subplot 1
plt.subplot(121)
plt.ylabel('Yup')
values = [wdf_nrt[wdf_nrt['RainTomorrow'] == 'Yes'].shape[0],wdf_rt[wdf_rt['RainTomorrow'] == 'Yes'].shape[0]]
plt.bar(names, values)
# subplot 2
plt.subplot(122)
plt.ylabel('Nope')
values = [wdf_nrt[wdf_nrt['RainTomorrow'] == 'No'].shape[0],wdf_rt[wdf_rt['RainTomorrow'] == 'No'].shape[0]]
plt.bar(names, values)
# set the overall title
plt.suptitle('Will it rain tomorrow?')
# add a gap
plt.tight_layout(w_pad=5)
# show all plots
plt.show()
```
